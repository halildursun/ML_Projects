<h2>Categorical Variables and One Hot Encoding and Pandas Get Dummies</h2>

```python
#import modules
import pandas as pd
```

```python
# read df
df = pd.read_csv("carprices.csv")
df
```

```python
#Explore Data
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel('Car Modul')
plt.ylabel('Price')
plt.scatter(df["Car Model"],df["Sell Price($)"],color='red',marker='+')
```

```python
#get dummies
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
#concat df
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
#drop column
final = merged.drop(["Car Model","Mercedez Benz C class"],axis='columns')
final
```

```python
# Features
X = final.drop('Sell Price($)',axis='columns')
X
```

```python
y = df["Sell Price($)"]
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
# predict price mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000,4,0,0]])
```

```python
# BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000,7,0,1]])
```

Using sklearn OneHotEncoder

First step is to use label encoder to convert town names into numbers

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle["Car Model"] = le.fit_transform(dfle["Car Model"])
dfle
```

```python
X = dfle.drop('Sell Price($)',axis='columns')
X
```

```python
y = dfle["Sell Price($)"]
y
```

Now use one hot encoder to create dummy variables for each of the town

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('"Car Model"', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,45000,2]])
```

```python
model.predict([[1,0,86000,7]])
```

```python
model.score(X,y)
```
